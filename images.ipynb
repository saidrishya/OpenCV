```python
import cv2
```

```python
frame = cv2.imread('superman.jpg')
plt.imshow(frame)
```

```python
frame_new = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(frame_new)
```

```python
classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face = classifier.detectMultiScale(frame)
faces = face[0]
x,y,w,h = faces
face = frame_new[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

```python
out = cv2.rectangle(frame_new, (x,y) , (x+w , y+h) ,(0,0,255) , 6)
plt.imshow(out)
```
